```
import numpy as np
import random

def create_chromosome():
    return {
        "stop_loss": random.uniform(1, 99),
        "take_profit": random.uniform(1, 99),
        "trade_size": random.uniform(1, 99)
    }

def initialize_population(size):
    chromosomes = []
    for i in range(size):
        chromosome = create_chromosome()
        chromosomes.append(chromosome)
    return chromosomes

def evaluate_fitness(chromosome,HISTORICAL_PRICES,initial_capital=1000):
    capital = initial_capital
    for i in HISTORICAL_PRICES:
        trade_size = capital * (chromosome["trade_size"] / 100)
        if i <= -chromosome["stop_loss"]:
            loss = trade_size * (chromosome["stop_loss"] / 100)
            capital -= loss
        elif i >= chromosome["take_profit"]:
            profit = trade_size * (chromosome["take_profit"] / 100)
            capital += profit

    return capital - initial_capital

def select_parents(population):
    return random.sample(population, 2)
#For Task-01
def crossover(parent1, parent2):
    child1 = parent1.copy()
    child2 = parent2.copy()
    crossover_point = random.choice(["stop_loss", "take_profit", "trade_size"])
    temp = child1[crossover_point]
    child1[crossover_point] = parent2[crossover_point]
    child2[crossover_point] = temp
    return child1, child2
#For Task-02_Start_Here_________________________________________________
def TwoPointerCrossOver(parent1, parent2):
    Genes = ["stop_loss", "take_profit", "trade_size"]
    crossover_points = random.sample(range(len(Genes)), 2)
    crossover_points.sort()
    child1 = parent1.copy()
    child2 = parent2.copy()
    for i in range(crossover_points[0], crossover_points[1] + 1):
        gene = Genes[i]
        temp = child1[gene]
        child1[gene] = parent2[gene]
        child2[gene] = temp
    return child1, child2
#For_Task_02_Ends_Here_____________________________________________________
def mutation(chromosome, mutation_rate):
    if random.random() < mutation_rate:
        gene = random.choice(["stop_loss", "take_profit", "trade_size"])
        new_value = random.uniform(1, 99)
        chromosome[gene] = new_value
    return chromosome

def genetic_algorithm(POPULATION_SIZE,GENERATIONS,MUTATION_RATE,HISTORICAL_PRICES):
    population = initialize_population(POPULATION_SIZE)

    for i in range(GENERATIONS):
        for j in population:
            j["fitness"] = evaluate_fitness(j,HISTORICAL_PRICES)

        population.sort(key=lambda x: x["fitness"], reverse=True)
        new_population = population[:2]

        while len(new_population) < POPULATION_SIZE:
            parents = select_parents(population)
            parent1 = parents[0]
            parent2 = parents[1]
            Task_01 = crossover(parent1, parent2)
            child1 = Task_01[0]
            child2 = Task_01[1]
            #---------------------------------------------------------
            Task_02 = TwoPointerCrossOver(parent1,parent2)#Task-02
            child1_for_Task_02 = Task_02[0]#Task-02
            child2_for_Task_02 = Task_02[1]#Task-02
            #print(child1_for_Task_02)#Task-02
            #print(child2_for_Task_02)#Task-02
            #---------------------------------------------------------
            mutation1 =mutation(child1, MUTATION_RATE)
            new_population.append(mutation1)
            if len(new_population) < POPULATION_SIZE:
                mutation2 = mutation(child2, MUTATION_RATE)
                new_population.append(mutation2)
        population = new_population

    best_strategy = max(population, key=lambda x: x["fitness"])
    final_profit = best_strategy["fitness"]
    best_strategy.pop("fitness")
    return best_strategy,final_profit

HISTORICAL_PRICES = [-1.2, 3.4, -0.8, 2.1, -2.5, 1.7, -0.3, 5.8, -1.1, 3.5]
POPULATION_SIZE = 10
GENERATIONS = 10
MUTATION_RATE = 0.05
best_strategy = genetic_algorithm(POPULATION_SIZE,GENERATIONS,MUTATION_RATE,HISTORICAL_PRICES)
print(f"Best Trading Strategy: {best_strategy[0]}\nfinal_profit: {best_strategy[1]}")
```
