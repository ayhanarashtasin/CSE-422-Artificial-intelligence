```python
#Task-01
import math
import random

def strength(x):
    return math.log2(x + 1) + x / 10

def calculate_utility(maxV, minV):
    i = random.choice([0, 1])
    factor = random.randint(1, 10)
    return strength(maxV) - strength(minV) + ((-1) ** i) * factor / 10

def alpha_beta_proning(depth, nodeIndex, maximizingPlayer,MaxV,MinV,alpha, beta):
    if depth == 5:
        return calculate_utility(MaxV,MinV)
    if maximizingPlayer ==True:
      best = float('-inf')
      left =alpha_beta_proning(depth + 1, nodeIndex * 2, False, MaxV, MinV, alpha, beta)
      best = max(best, left)
      alpha = max(alpha, best)
      if beta <= alpha:
          return best

      right = alpha_beta_proning(depth + 1, nodeIndex * 2 + 1, False, MaxV, MinV, alpha, beta)
      best = max(best, right)
      alpha = max(alpha, best)
      return best

    else:
      best = float('inf')
      left = alpha_beta_proning(depth + 1, nodeIndex * 2, True, MaxV, MinV, alpha, beta)
      best = min(best, left)
      beta = min(beta, best)
      if beta <= alpha:
          return best
      right = alpha_beta_proning(depth + 1, nodeIndex * 2 + 1, True, MaxV, MinV, alpha, beta)
      best = min(best, right)
      beta = min(beta, best)
      return best

start = int(input("Enter starting player for game 1 (0 for Carlsen, 1 for Caruana): "))
carlsen_strength = float(input("Enter base strength for Carlsen: "))
caruana_strength = float(input("Enter base strength for Caruana: "))
carlsen_wins = 0
caruana_wins = 0
draws = 0
no_of_round = 4
result = 0

for i in range(4):
    if (start + i) % 2 == 0:
        result = alpha_beta_proning(0, 0, True, carlsen_strength, caruana_strength, float('-inf'), float('inf'))
        player = "Magnus Carlsen (Max)"
    else:
        result = alpha_beta_proning(0, 0, True, caruana_strength, carlsen_strength, float('-inf'), float('inf'))
        player = "Fabiano Caruana (Max)"

    if result > 0:
        carlsen_wins += 1
        winner = player
    elif result < 0:
        caruana_wins += 1
        if player == "Magnus Carlsen (Max)":
            winner = "Fabiano Caruana (Min)"
        else:
            winner = "Magnus Carlsen (Min)"
    else:
        draws += 1
        winner = "Draw"

    print(f"Game {i+1} Winner: {winner} (Utility value: {result})")

print("Overall Results:")
print(f"Magnus Carlsen Wins: {carlsen_wins}")
print(f"Fabiano Caruana Wins: {caruana_wins}")
print(f"Draws: {draws}")

if carlsen_wins > caruana_wins:
    print("Overall Winner: Magnus Carlsen")
elif caruana_wins > carlsen_wins:
    print("Overall Winner: Fabiano Caruana")
else:
    print("Overall Winner: Draw")
```
